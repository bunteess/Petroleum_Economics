```python
import numpy_financial as npf
import matplotlib.pyplot as plt

# Inputs
initial_investment = 1000000 # in dollars
discount_rate = 0.10 # annual discount rate
production = [20000, 25000, 30000, 35000, 40000, 45000] # annual oil production in barrels
prices = [50, 55, 60, 65, 70, 75] # price of oil per barrel in dollars

# Calculation
cash_flows = [-initial_investment] + [production[i] * prices[i] for i in range(len(production))]
npv = npf.npv(discount_rate, cash_flows)

# Plotting positive and negative cash flows
positive_cash_flows = [c for c in cash_flows if c > 0]
negative_cash_flows = [c for c in cash_flows if c < 0]

plt.bar(list(range(len(positive_cash_flows))), positive_cash_flows, color='green', edgecolor='black')
plt.bar(list(range(len(negative_cash_flows))), [c for c in negative_cash_flows], color='red', edgecolor='black')
plt.ylim(-max(cash_flows), max(cash_flows))
plt.axhline(y=0, color='black')
plt.xlabel('Years')
plt.ylabel('Cash Flows (in dollars)')
plt.title('NPV Analysis for Oil Field Production')
plt.show()

# Output
print("Net present value:", npv)
```
